```python
import random
from mpl_toolkits.mplot3d import Axes3D
import math
import numpy as np
from BatAlgorithm import *
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def funEasom(d):
    res = 1
    for i in d:
        res = res * math.cos(i) * math.exp(-1.0 * np.sum((np.array(d) - math.pi * np.ones(np.array(d).shape)) ** 2))
    if len(d) % 2 == 0:
        res = -1 * res
    return res
```

```python
N=100
d = np.ndarray([N,2],dtype=float)
d[:,0]=np.multiply(np.array(range(0,N)),0.1)
d[:,1]=np.multiply(np.array(range(0,N)),0.1)

f = np.ndarray([N,N],dtype=float)

fig = plt.figure(figsize=(15,10))

ax = fig.gca(projection='3d')

# f[:] = funEasom(np.array([d[:,0],d[:,1]]))

for j in range(0,N):
    for i in range(0,N):
        f[i,j] = funEasom(np.array([d[i,0],d[j,1]]))
  

surf = ax.plot_surface(
# отмечаем аргументы и уравнение поверхности
d[:,0],d[:,1],f, 
# шаг прорисовки сетки
# - чем меньше значение, тем плавнее
# - будет градиент на поверхности
rstride = 1,
cstride = 1,
# цветовая схема plasma
cmap = cm.plasma)

# ax.plot_surface(d[:,0],d[:,1],f)
```
