Функция Изома:
$${\displaystyle f(x,y)=-\cos \left(x\right)\cos \left(y\right)\exp \left(-\left(\left(x-\pi \right)^{2}+\left(y-\pi \right)^{2}\right)\right)}$$
                                                    <div align=center> <img src="http://www.geatbx.com/ver_3_3/msh_easo_1_5.gif"> </div>

```python
def funEasom(d):
    res = 1
    for i in d:
        res = res * math.cos(i) * math.exp(np.sum(-1.0*(np.array(d) - math.pi * np.ones(np.array(d).shape)) ** 2))
    if len(d) % 2 == 0:
        res = -1 * res
    return res
```

Функция Шуберта:
<div align=center> <img src="https://www.sfu.ca/~ssurjano/shubert2.png"> <img src="https://www.sfu.ca/~ssurjano/shubert.png"> </div>


```python
def funShubert(d):
    res = 1
    for x in d:
        loc_res = 0
        for i in range(1,6):
            loc_res += i*np.cos((i+1)*x+i)
        res = res * loc_res
    return res
```

Функция Лангермана:

<div align=center> <img src="https://www.sfu.ca/~ssurjano/langer.png"> </div>


```python
def funLangerman(d):
    res = 1
    d = np.array(d)
#     res = np.exp((-1/math.pi)* np.sum((d - np.ones(np.array(d).shape)) ** 2))*np.cos(math.pi* np.sum((d - np.ones(np.array(d).shape)) ** 2))
    res = np.exp((-1/math.pi)* np.sum(d * d))*np.cos(math.pi* np.sum(d * d))
    return res
```

Функция Розенброка:
$${\displaystyle f({\boldsymbol {x}})=\sum _{i=1}^{n-1}\left[100\left(x_{i+1}-x_{i}^{2}\right)^{2}+\left(x_{i}-1\right)^{2}\right]}$$

<div align=center> <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/7e/Rosenbrock%27s_function_in_3D.pdf/page1-800px-Rosenbrock%27s_function_in_3D.pdf.jpg"> </div>

```python
def funRosenbroke(d):
    res = 0
    for i in range(0,np.shape(d)[0]-2):
        res += 100*(d[i+1]-d[i]**2)+(d[i]-1)**2
    return res
```
