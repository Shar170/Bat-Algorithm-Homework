```python
import random
from mpl_toolkits.mplot3d import Axes3D
import math
import numpy as np
from BatAlgorithm import *
import matplotlib.pyplot as plt
from matplotlib import cm
```

Функция Изома:
$${\displaystyle f(x,y)=-\cos \left(x\right)\cos \left(y\right)\exp \left(-\left(\left(x-\pi \right)^{2}+\left(y-\pi \right)^{2}\right)\right)}$$
                                                    <div align=center> <img src="http://www.geatbx.com/ver_3_3/msh_easo_1_5.gif"> </div>

```python
def funEasom(d):
    res = 1
    for i in d:
        res = res * math.cos(i) * math.exp(np.sum(-1.0*(np.array(d) - math.pi * np.ones(np.array(d).shape)) ** 2))
    if len(d) % 2 == 0:
        res = -1 * res
    return res
```

```python
dimensi= 3
n_bat=100
n_generasi=100000
r0=0.81
alpha=0.95
gamma=0.95
fmin=0
fmax=1
b_down =-10
b_up =10
maxCount = 700

ba_array = []

for i in range(0,10):
    ba_array.append(BatAlgorithm(dimensi, n_bat, n_generasi, r0, alpha-i/10.0, gamma, fmin, fmax, b_down, b_up, funEasom, maxCount))

for i in range(0,10):
    ba_array[i].proses_ba()
    print('alpha= ', alpha-i/10.0)

fig, ax = plt.subplots(figsize=(7,6))
ax.set_yscale('log')

for i in range(0,10):
    plt.plot(np.array(ba_array[i].history)+1, label=f'alpha={alpha-i/10.0}')

plt.legend()
```
